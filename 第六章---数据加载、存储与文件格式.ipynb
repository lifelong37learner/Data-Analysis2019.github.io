# 利用lxml.ibjectify解析XML

```python
from lxml import objectify
```

```python
from lxml import objectify

path = r'C:\Users\Administrator\Desktop\新一期DataAnalysis\ch06\mta_perf\Performance_MNR.xml'
parsed = objectify.parse(open(path))
root = parsed.getroot()
```

```python
data = []
skip_fields = ['PARENT_REQ', 'INDICATOR_SEQ', "DESIRED_CHANGE", 'DECIMAL_PLACES']
```

```python
# root.INDICATOR 返回一个用于产生各个<INDICATOR> XML元素的生成器
for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

```python
pref = DataFrame(data)
pref
```

# 使用数据库

```python
import sqlite3
```

```python
query = """
    CREATE TABLE test (a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER);
"""
con = sqlite3.connect(':memory:')
con.execute(query)
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
       ('Tallahassee', 'Florida', 2.6, 3),
       ('Sacramento', 'California', 1.7, 5)]
```

```python
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"

con.executemany(stmt, data)
con.commit()
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall()
rows
```

```python
# 可以将这个元祖列表穿个DataFrame的构造器，但是还需要列名（位于游标的description属性中）
cursor.description
```

```python
frame3 = DataFrame(rows, columns =list(zip(*cursor.description))[0]) #zip外面再套一层list，即使用list包装zip对象！
```
